```python
# Get unique combinations of make, model, and modality (from auxo)
auxo_df = auxo_df[['MAKE','MODEL']]
auxo_df = auxo_df.drop_duplicates()
auxo_df
```

```python
# Add standardized make (from prior script)
raw_df = pd.merge(raw_df, std_make_df, left_on='make', right_on='make_source', how='left')
raw_df
```

```python
if TARGET_FILE == 'auxo':
    target_df = auxo_df
    manufacturer_col = 'MAKE'
    model_col = 'MODEL'
elif TARGET_FILE == 'mel':
    target_df = mel_df
    manufacturer_col = 'Manufacturer'
    model_col = 'Model'
elif TARGET_FILE == 'uptime':
    target_df = uptime_df
    manufacturer_col = 'MAKE_GOLDEN'
    model_col = 'MODEL_GOLDEN'

x = pd.merge(raw_df, target_df, left_on = ['make_mel','model_name'], right_on = [manufacturer_col, model_col], how='left')
x
```
